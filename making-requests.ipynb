How to make a conversation with Claude

````python
# Helper fucntion for maintatining history

# Append user message to messages list
def add_user_message(messages, text):
    user_message = { "role": "user", "content": text}
    messages.append(user_message)

# Append assistant message to messages list
def add_assistant_message(messages, text):
    assistant_message = { "role": "assistant", "content": text}
    messages.append(assistant_message)

# Chat function
def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1024,
        messages=messages
    )
    return message.content[0].text
````

````python
# Idea is to maintain a list of messages and pass it to the chat function
# Starting list of messages
messages = []

# Add initial user message
add_user_message(messages, "What is quantum computing? Answer in a sentence.")

# Get assistant response
answer = chat(messages)

# Add assistant response to messages
add_assistant_message(messages, answer)

# Add naother user message
add_user_message(messages, "Write another sentence.")

answer = chat(messages)
answer
````

Create a countinous conversation with Claude

````python
messages = []

while True:
    user_input = input("> ")
    print("> ", user_input)
    add_user_message(messages, user_input)
    # Call claude to get answer
    answer = chat(messages)

    # Add asisstant message to history
    add_assistant_message(messages, answer)
    print("----")
    print("Claude: " + answer)
    print("----")
````

````python
messages = []

add_user_message(messages, "How do I solve 5x+3=2 for x?")

system_prompt = """You are a patient math tutor.
Do not directly answer the question. Instead, guide the student through the solution step by step.
"""

# Chat function
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1024,
        "messages": messages
    }
    if system:
        params["system"] = system
    message = client.messages.create(
        **params
    )
    return message.content[0].text

answer = chat(messages, system=system_prompt)
answer
````

Response streaming

````python
messages = []

add_user_message(messages, "Write a 1 sentecne description about a fake database.")

stream = client.messages.create(
    model=model,
    max_tokens=1024,
    messages=messages,
    stream=True
)

for event in stream:
    print(event)
````

````python
messages = []
add_user_message(messages, "Write a 1 sentecne description about a fake database.")

with client.messages.stream(
    model=model,
    max_tokens=1024,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end='')
````

Stop Sequences

````python
messages = []
add_user_message(messages, "List me 1 to 10 numbers.")
answer = chat(messages)
answer
````

````python
# Chat function
def chat(messages, system=None, temperature=1.0, stop_sequences=[]):
    params = {
        "model": model,
        "max_tokens": 1024,
        "messages": messages, 
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }
    if system:
        params["system"] = system
    message = client.messages.create(
        **params
    )
    return message.content[0].text

messages = []
add_user_message(messages, "List me 1 to 10 numbers.")
answer = chat(messages, stop_sequences=["5"])
answer
````

Structured Data

When you need Claude to generate structured data like JSON, Python code, or bulleted lists, you'll often run into a common problem: Claude wants to be helpful and add explanatory text around your content. While this is usually great, sometimes you need just the raw data with nothing else.

````python
messages = []

add_user_message(messages, "Write a json imaginary adress of John Doe in json format.")
answer = chat(messages)
answer
````
